# Create the target table

```python
%sql
CREATE OR REPLACE TABLE workspace.silver.erp_loc_a101
(customer_id STRING,
country STRING
) USING DELTA
```

# Do the Transformations in the original table & insert to the target table

### Column Parsing & Normalization

```python
%sql
INSERT OVERWRITE workspace.silver.erp_loc_a101
SELECT 
REPLACE(CID, '-','') AS customer_id,
CNTRY AS country
 FROM workspace.bronze.erp_loc_a101
```
