# Drop the target table if already exists

# Create the target table

```python
%sql
CREATE OR REPLACE TABLE workspace.silver.erp_cust_az12
(
  customer_number STRING,
  birthdate DATE,
  gender STRING) USING DELTA
```

# Do the Transformations in the original table & insert to the target table

### Column Parsing & Normalization

```python
%sql
INSERT OVERWRITE workspace.silver.erp_cust_az12
SELECT SUBSTRING(CID, 4, LEN(CID)) AS customer_number,
BDATE AS birthdate,
CASE
  WHEN UPPER(TRIM(GEN)) = 'M' THEN 'Male'
  WHEN UPPER(TRIM(GEN)) = 'F' THEN 'Female'
  WHEN UPPER(TRIM(GEN)) = 'MALE' THEN 'Male'
  WHEN UPPER(TRIM(GEN)) = 'FEMALE' THEN 'Female'
  ELSE 'n/a'
  END AS gender
FROM workspace.bronze.erp_cust_az12
```
