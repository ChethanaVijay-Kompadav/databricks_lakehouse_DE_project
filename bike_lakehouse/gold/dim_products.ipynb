```python
%sql
DROP TABLE IF EXISTS workspace.gold.dim_products
```

```python
%sql
CREATE OR REPLACE TABLE workspace.gold.dim_products(
product_key INT,
  product_id INT,
  category_id STRING,
  product_number STRING,
  product_name STRING,
  category STRING,
  subcategory STRING,
  maintenance STRING,
  cost INT,
  product_line STRING,
  start_date DATE,
  end_date  DATE) USING DELTA
```

```python
%sql
SELECT * FROM workspace.gold.dim_products
```

```python
%sql
INSERT OVERWRITE workspace.gold.dim_products
SELECT 
  ROW_NUMBER()OVER(ORDER BY pi.start_date, pi.product_number) AS product_key,
  pi.product_id,
  pi.category_id,
  pi.product_number,
  pi.product_name,
  pc.category,
  pc.subcategory,
  pc.maintenance,
  pi.cost,
  pi.product_line,
  pi.start_date,pi.end_date
FROM workspace.silver.crm_products pi
LEFT JOIN workspace.silver.erp_px_cat_g1v2 pc
ON pi.category_id = pc.category_id
WHERE pi.end_date IS NULL
```
