```python
%sql
DROP TABLE IF EXISTS workspace.gold.fact_sales
```

```python
%sql
CREATE OR REPLACE TABLE workspace.gold.fact_sales(
  order_number STRING,
  product_key INT,
  customer_id INT,
  order_date DATE,
  ship_date DATE,
  due_date DATE,
  sales_amount INT,
  quantity INT,
  price INT
) USING DELTA
```

```python
%sql
SELECT * FROM workspace.gold.fact_sales
```

```python
%sql
INSERT OVERWRITE workspace.gold.fact_sales
SELECT
  s.order_number,
  p.product_key,
  c.customer_id,
  s.order_date,
  s.ship_date,
  s.due_date,
  s.sales_amount,
  s.quantity,
  s.price
FROM workspace.silver.crm_sales s
LEFT JOIN workspace.gold.dim_products p
ON s.product_number = p.product_number
LEFT JOIN workspace.gold.dim_customers c
ON s.customer_number = c.customer_id
```
